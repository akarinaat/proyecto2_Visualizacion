```python
import pandas as pd
from pandasql import sqldf
```

```python
#Leemos el excel del 2022
df = pd.read_excel("DATOS_ABIERTOS_JULIO_2022.xlsx")

#Modificamos la fecha para facilitar su uso
df['FECHA_INGRESO'] = df['FECHA_INGRESO'].dt.strftime('%Y-%m-%d')
#Renombramos algunas tablas para facilitar su uso
df = df.rename(columns={"MONTO RECLAMADO":"MONTO_RECLAMADO","MONTO RECUPERADO":"MONTO_RECUPERADO"})
#Selecionamos las tablas de nuestro interes
df = sqldf("SELECT SECTOR,TIPO_RECLAMACION,ESTADO_PROCESAL,COSTO_BIEN,MONTO_RECLAMADO,MONTO_RECUPERADO FROM df")
#Llenamos los nan con 0
df = df.fillna(0)

#Despliegamos el df
df
```

```python
#Leemos el excel del 2021
df2 = pd.read_excel("DATOS_ABIERTOS_2021_2.xlsx")

#Modificamos la fecha para facilitar su uso
df2['FECHA_INGRESO'] = df2['FECHA_INGRESO'].dt.strftime('%Y-%m-%d')
#Selecionamos las fechas de Enero-Julo
df2 = df2.loc[(df2['FECHA_INGRESO'] < '2021-08-01')]
#Renombramos algunas tablas para facilitar su uso
df2 = df2.rename(columns={"RECLAMADO":"MONTO_RECLAMADO","RECUPERADO*":"MONTO_RECUPERADO"})
#Selecionamos las tablas de nuestro interes
df2 = sqldf("SELECT SECTOR,TIPO_RECLAMACION,ESTADO_PROCESAL,COSTO_BIEN,MONTO_RECLAMADO,MONTO_RECUPERADO FROM df2")
#Llenamos los nan con 0
df2 = df2.fillna(0)

#Despliegamos el df
df2
```

# Incidencia del costo del bien y la queja

```python
#Tranformamos los objetos a flotantes
df["COSTO_BIEN"] = pd.to_numeric(df['COSTO_BIEN'],errors='coerce')
df2["COSTO_BIEN"] = pd.to_numeric(df2['COSTO_BIEN'],errors='coerce')
df2["MONTO_RECLAMADO"] = pd.to_numeric(df2['MONTO_RECLAMADO'],errors='coerce')
df2["MONTO_RECUPERADO"] = pd.to_numeric(df2['MONTO_RECUPERADO'],errors='coerce')
```

```python
#Creamos Rangos del costo de bien y vemos su promedio de monto recuperado
#Para ver si vale la pane hacer el reclamo
# incidencias 2022
incidencia_2022 = sqldf('''SELECT
            CASE
                WHEN COSTO_BIEN < 1000 THEN "0 - 999"
                WHEN COSTO_BIEN < 10000 THEN "1,000 - 9,999"
                WHEN COSTO_BIEN < 100000 THEN "10,000 - 99,999"
                ELSE "100,000 +"
            END 
            AS COSTO_BIEN_RANGE,COUNT(*) AS FREQUENCIA, AVG(MONTO_RECUPERADO) AS AVG_MONTO_RECUPERADO
            FROM df WHERE MONTO_RECUPERADO != 0 GROUP BY COSTO_BIEN_RANGE''')
incidencia_2022
```

```python
#Creamos Rangos del costo de bien y vemos su promedio de monto recuperado
#Para ver si vale la pane hacer el reclamo
# incidencias 2021
incidencia_2021 = sqldf('''SELECT
            CASE
                WHEN COSTO_BIEN < 1000 THEN "0 - 999"
                WHEN COSTO_BIEN < 10000 THEN "1,000 - 9,999"
                WHEN COSTO_BIEN < 100000 THEN "10,000 - 99,999"
                ELSE "100,000 +"
            END 
            AS COSTO_BIEN_RANGE,COUNT(*) AS FREQUENCIA, AVG(MONTO_RECUPERADO) AS AVG_MONTO_RECUPERADO
            FROM df2 WHERE MONTO_RECUPERADO != 0 GROUP BY COSTO_BIEN_RANGE''')
incidencia_2021
```
